```python
# Install library (only needed in Google Colab)
!pip install yfinance

# Import libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Select stocks
stocks = ['RELIANCE.NS', 'TCS.NS', 'HDFCBANK.NS']

# Step 2: Download stock data
data = yf.download(stocks, start='2022-01-01', end='2024-01-01')['Close']

# Step 3: Show first few rows
print("Stock Data:")
print(data.head())

# Step 4: Plot stock price trends
plt.figure(figsize=(12,6))
for stock in stocks:
    plt.plot(data[stock], label=stock)

plt.title("Stock Price Trend Analysis")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()

# Step 5: Calculate Moving Averages
ma20 = data.rolling(window=20).mean()
ma50 = data.rolling(window=50).mean()

# Plot Moving Average for one stock
plt.figure(figsize=(12,6))
plt.plot(data['RELIANCE.NS'], label='Actual Price')
plt.plot(ma20['RELIANCE.NS'], label='20-Day MA')
plt.plot(ma50['RELIANCE.NS'], label='50-Day MA')

plt.title("Moving Average - RELIANCE")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()

# Step 6: Daily Returns
returns = data.pct_change()

plt.figure(figsize=(12,6))
returns.plot(title="Daily Returns")
plt.show()

# Step 7: Volatility
volatility = returns.std() * (252**0.5)
print("\nVolatility (Risk):")
print(volatility)
```
